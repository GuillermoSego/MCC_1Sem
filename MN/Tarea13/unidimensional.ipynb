```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_solution(filename):
    return np.loadtxt(filename)

def y_real(t):
    return np.exp(t)

# Cargar soluciones
euler = load_solution("Euler.txt")
heun = load_solution("Heun.txt")
taylor = load_solution("Taylor2.txt")
rk4 = load_solution("RK4.txt")
```

```python
# Generar puntos t
a = 0  # Punto inicial
b = 4  # Punto final
N = 10  # Número de pasos
t = np.linspace(a, b, N + 1)

# Solución real
# real_solution = y_real(t)
```

```python
# Graficar
plt.figure(figsize=(10, 6))
plt.plot(t, euler, label='Euler', marker='o')
plt.plot(t, heun, label='Heun', marker='x')
plt.plot(t, taylor, label='Taylor', marker='^')
plt.plot(t, rk4, label='RK4', marker='s')
# plt.plot(t, real_solution, label='Real (e^x)', linestyle='dashed')

plt.xlabel('t')
plt.ylabel('y')
plt.title('Comparación de Métodos de Resolución de EDOs')
plt.legend()
plt.grid(True)
plt.savefig("Problema2.pdf")
plt.show()
```
