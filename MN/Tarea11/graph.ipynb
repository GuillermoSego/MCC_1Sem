```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Función original de Sutherland
def sutherland(T, m0, T0, Su):
    return m0 *  ((1000*T)/T0)**(3/2) * (T0 + Su) / (1000*T + Su)

# Funciones de ajuste
def polinomial(x, coefs):
    return sum(a * x**i for i, a in enumerate(coefs))

def trigonometric(x, coefs):
    return sum(a * np.cos(i * np.pi * x / 6) for i, a in enumerate(coefs))

def radial_basis_function(x, coefs, centers):
    return sum(a * np.exp(-((x - centers[i])**2)) for i, a in enumerate(coefs))

# Datos para la función original de Sutherland
m0 = 1.919e-2
T0 = 273
Su = 139
```

```python
# Coeficientes dados para los tres ajustes
coef_pol = [0.003811, 0.061038, -0.014866]
coef_trig = [0.049016, 0.020504, -0.048746]
coef_rbf = [-1.437938, 0.163605, 1.289774]

# Puntos xi para RBF (usamos los mismos puntos)
xi = [0.273, 0.303, 0.323, 0.353, 0.423, 0.573, 1.473]

# Rango de x para graficar
x_range = np.arange(0.273, 1.5, 0.05)

# Evaluación de las funciones
y_original = sutherland(x_range, m0, T0, Su)
y_pol = polinomial(x_range, coef_pol)
y_trig = trigonometric(x_range, coef_trig)
y_rbf = radial_basis_function(x_range, coef_rbf, xi)

# Grafica
plt.figure(figsize=(10, 8))

plt.plot(x_range, y_original, label='Función Original Sutherland', color='black')
plt.plot(x_range, y_pol, label='Ajuste Polinomial', linestyle='--', color='red')
plt.plot(x_range, y_trig, label='Ajuste Trigonométrico', linestyle='--', color='green')
plt.plot(x_range, y_rbf, label='Función RBF', linestyle='--', color='blue')

plt.title('Comparación de la función de S(T) y funciones de ajuste para $\lambda$ = 1.000000e-05 con m = 3')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.savefig('l5m3.pdf')
plt.show()
```

```python
# Coeficientes actualizados para los tres ajustes con diferentes m
coeficientes = {
    'm2': {
        'pol': [0.011386, 0.034514],
        'trig': [0.166103, -0.144770],
        'rbf': [-2.358377, 2.372265]
    },
    'm3': {
        'pol': [0.003799, 0.061078, -0.014888],
        'trig': [0.049017, 0.020504, -0.048748],
        'rbf': [-1.459877, 0.165949, 1.309193]
    },
    'm4': {
        'pol': [0.007295, 0.041021, 0.017930, -0.014129],
        'trig': [-0.066033, 0.042951, 0.174006, -0.135672],
        'rbf': [-0.786236, -0.213104, 0.190246, 0.826662]
    }
}

# Puntos xi para RBF
xi = [0.273, 0.303, 0.323, 0.353, 0.423, 0.573, 1.473]

# Rango de x para graficar
x_range = np.arange(0.273, 1.5, 0.05)

# Evaluar función original Sutherland
y_original = sutherland(x_range, m0, T0, Su)

# Crear la figura y el eje
plt.figure(figsize=(10, 8))

# Graficar la función original Sutherland
plt.plot(x_range, y_original, label='Función Original Sutherland', color='black', linewidth=2)

# Graficar los ajustes para cada m
for m, coefs in coeficientes.items():
    y_pol = polinomial(x_range, coefs['pol'])
    y_trig = trigonometric(x_range, coefs['trig'])
    y_rbf = radial_basis_function(x_range, coefs['rbf'], xi)
    
    plt.plot(x_range, y_pol, label=f'Ajuste Polinomial m={m[1:]}', linestyle='--')
    plt.plot(x_range, y_trig, label=f'Ajuste Trigonométrico m={m[1:]}', linestyle='-.')
    plt.plot(x_range, y_rbf, label=f'Función RBF m={m[1:]}', linestyle=':')

# Configurar título y leyendas
plt.title('Comparación de la función de S(T) y funciones de ajuste para diferentes m')
plt.xlabel('Temperatura (T)')
plt.ylabel('Viscosidad (S)')
plt.legend()
plt.grid()

# Guardar la figura
plt.savefig('comparacion_ajustes.pdf')

# Mostrar la figura
plt.show()
```
