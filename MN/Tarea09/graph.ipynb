```python
import numpy as np
import matplotlib.pyplot as plt

# Valores experimentales
I_values = np.array([0.001, 0.005, 0.010, 0.015, 0.019, 0.022])
V_values = np.array([0.7252, 0.62625, 0.57969, 0.53438, 0.43906, 0.28125])

# Leer los valores interpolados desde el archivo
I_interpolated, V_interpolated = np.loadtxt('interpolated_values.txt', unpack=True)

# Graficando la curva de polarización y valores interpolados
plt.plot(I_values, V_values, 'bo-', label='Datos experimentales')
plt.plot(I_interpolated, V_interpolated, 'ro', label='Valores interpolados')
plt.xlabel('I (A/cm^2)')
plt.ylabel('Voltage (V)')
plt.title('Curva de Polarización')
plt.legend()
plt.grid()
plt.savefig('graph.pdf')
plt.show()
```
