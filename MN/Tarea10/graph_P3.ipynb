```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Leer datos
def leer_datos(nombre_archivo):
    return np.loadtxt(nombre_archivo, unpack=True)

x1, f1 = leer_datos("Curva1_P3.txt")
x2, f2 = leer_datos("Curva2_P3.txt")
x3, f3 = leer_datos("Curva3_P3.txt")
```

```python
# Grfica
plt.figure(figsize=(20, 8))

# Grfica de las curvas
plt.plot(x1, f1, label='Curva 1', color='b')
plt.plot(x2, f2, label='Curva 2', color='brown')
plt.plot(x3, f3, label='Curva 3', color='black')

# Ajustar los lmites de los ejes
plt.xlim(0, 32)
plt.ylim(0, 10)

# Etiquetas y leyenda
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Grfica del perrito')
plt.legend()
plt.grid()
plt.savefig("perrito.pdf")
plt.show()
```
