```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Leer datos
x, f = np.loadtxt("DataP2.txt", unpack=True)
```

```python
# Gráfica
plt.figure(figsize=(20,8))

# Gráfica de la curva
plt.plot(x, f, label='Curva Pato')

# Ajustar los límites de los ejes
plt.xlim(0, 14)
plt.ylim(0, 5)

# Etiquetas y leyenda
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfica del pato')
plt.legend()
plt.grid()
plt.savefig("pato.pdf")
plt.show()

```
