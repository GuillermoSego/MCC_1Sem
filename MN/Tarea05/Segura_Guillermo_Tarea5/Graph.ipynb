```python
# Importamos la biblioteca matplotlib para la graficación
import matplotlib.pyplot as plt

# Función para leer las temperaturas desde un archivo
def read_temps(filename):
    f = open(filename, 'r')
    lines = f.readlines()
    f.close()

    temps = []
    for x in lines:
        stripped = x.strip() # Quitar espacios en blanco y caracteres de nueva línea
        number = float(stripped) # Convertir a flotante
        temps.append(number) # Añadir a la lista temps

    return temps

# Definir la lista de n y los nombres de los archivos correspondientes
n_values = [10, 30, 50, 70, 100]
files = []
for n in n_values:
    file_name = "n" + str(n) + ".txt"
    files.append(file_name)

# Lista para almacenar las temperaturas centrales
central_temps = []

# Leer las temperaturas del nodo central para cada archivo
for filename in files:
    temps = read_temps(filename) # Leer temperaturas del archivo
    middle_index = len(temps) // 2 # Calcular el índice del medio
    central_temp = temps[middle_index] # Tomar la temperatura del nodo central
    central_temps.append(central_temp) # Añadir a la lista


# Grafica
plt.figure(figsize=(10, 6))
plt.plot(n_values, central_temps, marker='o')
plt.xlabel("Número de elementos (n)")
plt.ylabel("Temperatura del nodo central")
plt.title("Variación de la Temperatura del Nodo Central en función de n")
plt.grid()

plt.savefig('Tvsn.pdf')
plt.show()
```
