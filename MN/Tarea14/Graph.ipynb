```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Leer los datos del archivo
filename = 'output.txt'
data = np.loadtxt(filename)

# Preparar la malla para el gráfico
nt, nx = data.shape
x = np.linspace(0, 1, nx)  # Ajusta esto según tu domipnio espacial
t = np.linspace(0, 0.4, nt)  # Ajusta esto según tu dominio temporal

T, X = np.meshgrid(t, x)

# Graficar
plt.figure(figsize=(10, 6))
plt.contourf(T, X, data.T, 20, cmap='hot') 
plt.colorbar(label='Temperatura')
plt.xlabel('Tiempo')
plt.ylabel('Posición en la barra')
plt.title('Evolución de la Temperatura en la Barra de Calor')
plt.savefig('HeatTime.pdf')
plt.show()
```
